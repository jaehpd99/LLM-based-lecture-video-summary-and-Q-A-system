```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core.node_parser import SimpleNodeParser
from llama_index.core.response_synthesizers import ResponseMode

# Hugging Face 인증 토큰
hf_token = ""
model_name = "naver-hyperclovax/HyperCLOVAX-SEED-Text-Instruct-1.5B"

# 모델 및 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(model_name, token=hf_token)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    token=hf_token,
    device_map="auto",
    torch_dtype="auto"
)

# 패드 토큰 설정
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# 더 강력한 프롬프트 템플릿 설정
context_prompt = """당신은 주어진 문서 내용만을 바탕으로 답변하는 AI입니다.

중요한 규칙:
1. 반드시 아래 제공된 문서 내용만을 사용하여 답변하세요
2. 문서에 없는 내용은 절대 추가하지 마세요
3. 문서에서 답을 찾을 수 없으면 "제공된 문서에서 해당 정보를 찾을 수 없습니다"라고 답하세요
4. 일반적인 지식이나 사전 학습된 정보를 사용하지 마세요
5. 같은 내용을 반복하지 마세요
6. 간결하고 명확하게 답변하세요

문서 내용:
{context_str}

질문: {query_str}

답변 (간결하게):"""

# LLM 래퍼 설정 (반복 방지 파라미터 추가)
llm = HuggingFaceLLM(
    model=model,
    tokenizer=tokenizer,
    model_name=model_name,
    tokenizer_name=model_name,
    context_window=2048,
    max_new_tokens=256,
    query_wrapper_prompt=context_prompt,
    generate_kwargs={
        "do_sample": True,
        "temperature": 0.3,
        "top_p": 0.8,
        "repetition_penalty": 1.2,  # 반복 방지
        "no_repeat_ngram_size": 3,  # 3-gram 반복 방지
        "pad_token_id": tokenizer.pad_token_id,
        "eos_token_id": tokenizer.eos_token_id
    }
)

# 임베딩 모델 설정
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")

# LlamaIndex 설정
Settings.llm = llm
Settings.embed_model = embed_model

# 문서 로딩 및 노드 파싱 (더 작은 청크 크기로 정확도 향상)
documents = SimpleDirectoryReader(input_files=["/content/transcript.txt"]).load_data()

# 문서 내용 확인
print("📄 로드된 문서 정보:")
print(f"문서 수: {len(documents)}")
for i, doc in enumerate(documents):
    print(f"문서 {i+1} 길이: {len(doc.text)} 문자")
    print(f"문서 {i+1} 미리보기: {doc.text[:200]}...")

parser = SimpleNodeParser.from_defaults(
    chunk_size=400,  # 더 작은 청크로 정확도 향상
    chunk_overlap=100  # 더 큰 오버랩으로 정보 손실 방지
)
nodes = parser.get_nodes_from_documents(documents)

print(f"생성된 노드 수: {len(nodes)}")

# 벡터 인덱스 생성
index = VectorStoreIndex(nodes)

# 질의엔진 생성 (더 많은 컨텍스트 검색)
query_engine = index.as_query_engine(
    similarity_top_k=13,  # 더 많은 관련 청크 검색
    response_mode=ResponseMode.COMPACT,  # 컴팩트 모드로 더 집중된 답변
    streaming=False
)

# 🖥️ 인터페이스 시작
print("\n🎓 강의 영상 질의응답 시스템이 준비되었습니다!")
print("=" * 55)

# 초기 요약 생성 (더 구체적인 질문으로)
initial_query = "이 문서에서 다루는 주요 주제 3가지만 간단히 말해줘"
print("📋 강의 요약을 생성중...")
summary_response = query_engine.query(initial_query)
print("📝 **답변:**")
print(summary_response.response)

# 검색된 소스 정보 확인
'''
print("\n🔍 **검색된 소스 정보:**")
if hasattr(summary_response, 'source_nodes'):
    for i, node in enumerate(summary_response.source_nodes):
        print(f"소스 {i+1}: {node.text[:100]}...")
        if hasattr(node, 'score'):
            print(f"유사도 점수: {node.score:.3f}")
'''

print("\n" + "=" * 55)
print("\n💡 사용 가능한 명령어:")
print("- 질문을 입력하세요 (예: '주요 개념이 뭐야?')")
print("- 'debug': 마지막 답변의 소스 확인")
print("- 'history': 대화 기록 보기")
print("- 'summary': 강의 요약 다시 보기")
print("- 'check_doc': 문서 내용 일부 확인")
print("- 'exit': 종료")
print("-" * 55)

# 반복 제거 함수
def clean_response(text):
    """반복된 문장 제거"""
    sentences = text.split('.')
    unique_sentences = []
    seen = set()

    for sentence in sentences:
        sentence = sentence.strip()
        if sentence and sentence not in seen and len(sentence) > 10:
            unique_sentences.append(sentence)
            seen.add(sentence)

    return '. '.join(unique_sentences)

# 질의응답 루프
history = []
last_response = None

def show_source_info(response):
    """응답의 소스 정보를 표시하는 함수"""
    if hasattr(response, 'source_nodes') and response.source_nodes:
        print("\n🔍 **참조된 문서 부분:**")
        for i, node in enumerate(response.source_nodes):
            print(f"\n소스 {i+1}:")
            print(f"내용: {node.text[:200]}...")
            if hasattr(node, 'score'):
                print(f"유사도: {node.score:.3f}")
    else:
        print("\n⚠️ 소스 정보를 찾을 수 없습니다.")

while True:
    user_input = input("\n❓ 질문을 입력하세요: ").strip()

    if user_input.lower() == "exit":
        print("👋 질의응답을 종료합니다!")
        break

    elif user_input.lower() == "debug":
        if last_response:
            show_source_info(last_response)
        else:
            print("⚠️ 아직 질문이 없습니다.")

    elif user_input.lower() == "summary":
        print("\n📋 강의 요약:")
        print(clean_response(summary_response.response))
        show_source_info(summary_response)

    elif user_input.lower() == "history":
        print("\n📜 대화 기록:")
        for i, (q, a) in enumerate(history):
            print(f"{i+1}. Q: {q}")
            print(f"   A: {a[:200]}...")
            print()

    elif user_input.lower() == "check_doc":
        print("\n📄 문서 내용 확인:")
        if nodes:
            print(f"첫 번째 노드 내용: {nodes[0].text[:300]}...")
        else:
            print("노드가 없습니다.")

    elif user_input:
        # 질문 앞에 문서 기반 답변 강제하는 프리픽스 추가
        enhanced_query = f"문서 내용을 바탕으로만 답변해주세요: {user_input}"
        response = query_engine.query(enhanced_query)
        last_response = response

        # 반복 제거 처리
        cleaned_answer = clean_response(response.response)

        history.append((user_input, cleaned_answer))
        print("📝 **답변:**")
        print(cleaned_answer)

        # 소스 정보 표시
        # show_source_info(response)

    else:
        print("❗ 질문을 입력해주세요.")

# 추가 디버깅 함수들
def test_retrieval(query, top_k=3):
    """검색 기능 테스트"""
    retriever = index.as_retriever(similarity_top_k=top_k)
    nodes = retriever.retrieve(query)
    print(f"\n🔍 '{query}' 검색 결과:")
    for i, node in enumerate(nodes):
        print(f"\n결과 {i+1}:")
        print(f"내용: {node.text[:200]}...")
        if hasattr(node, 'score'):
            print(f"점수: {node.score:.3f}")

# 사용 예시:
# test_retrieval("주요 내용")
```

```python
# Hugging Face 인증 토큰
hf_token = ""
model_name = "naver-hyperclovax/HyperCLOVAX-SEED-Text-Instruct-1.5B"

# 모델 및 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(model_name, token=hf_token)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    token=hf_token,
    device_map="auto",
    torch_dtype="auto"
)

# 패드 토큰 설정
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# 더 강력한 프롬프트 템플릿 설정
context_prompt = """당신은 주어진 문서 내용만을 바탕으로 답변하는 AI입니다.

중요한 규칙:
1. 반드시 아래 제공된 문서 내용만을 사용하여 답변하세요
2. 문서에 없는 내용은 절대 추가하지 마세요
3. 문서에서 답을 찾을 수 없으면 "제공된 문서에서 해당 정보를 찾을 수 없습니다"라고 답하세요
4. 일반적인 지식이나 사전 학습된 정보를 사용하지 마세요
5. 같은 내용을 반복하지 마세요
6. 간결하고 명확하게 답변하세요

문서 내용:
{context_str}

질문: {query_str}

답변 (간결하게):"""

# LLM 래퍼 설정 (반복 방지 파라미터 추가)
llm = HuggingFaceLLM(
    model=model,
    tokenizer=tokenizer,
    model_name=model_name,
    tokenizer_name=model_name,
    context_window=2048,
    max_new_tokens=256,
    query_wrapper_prompt=context_prompt,
    generate_kwargs={
        "do_sample": True,
        "temperature": 0.3,
        "top_p": 0.8,
        "repetition_penalty": 1.2,  # 반복 방지
        "no_repeat_ngram_size": 3,  # 3-gram 반복 방지
        "pad_token_id": tokenizer.pad_token_id,
        "eos_token_id": tokenizer.eos_token_id
    }
)

# 임베딩 모델 설정
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")

# LlamaIndex 설정
Settings.llm = llm
Settings.embed_model = embed_model

# 문서 로딩 및 노드 파싱 (더 작은 청크 크기로 정확도 향상)
documents = SimpleDirectoryReader(input_files=["/content/transcript2.txt"]).load_data()

# 문서 내용 확인
print("📄 로드된 문서 정보:")
print(f"문서 수: {len(documents)}")
for i, doc in enumerate(documents):
    print(f"문서 {i+1} 길이: {len(doc.text)} 문자")
    print(f"문서 {i+1} 미리보기: {doc.text[:200]}...")

parser = SimpleNodeParser.from_defaults(
    chunk_size=400,  # 더 작은 청크로 정확도 향상
    chunk_overlap=100  # 더 큰 오버랩으로 정보 손실 방지
)
nodes = parser.get_nodes_from_documents(documents)

print(f"생성된 노드 수: {len(nodes)}")

# 벡터 인덱스 생성
index = VectorStoreIndex(nodes)

# 질의엔진 생성 (더 많은 컨텍스트 검색)
query_engine = index.as_query_engine(
    similarity_top_k=13,  # 더 많은 관련 청크 검색
    response_mode=ResponseMode.COMPACT,  # 컴팩트 모드로 더 집중된 답변
    streaming=False
)

# 🖥️ 인터페이스 시작
print("\n🎓 강의 영상 질의응답 시스템이 준비되었습니다!")
print("=" * 55)

# 초기 요약 생성 (더 구체적인 질문으로)
initial_query = "이 문서에서 다루는 주요 주제 3가지만 간단히 말해줘"
print("📋 강의 요약을 생성중...")
summary_response = query_engine.query(initial_query)
print("📝 **답변:**")
print(summary_response.response)

# 검색된 소스 정보 확인
'''
print("\n🔍 **검색된 소스 정보:**")
if hasattr(summary_response, 'source_nodes'):
    for i, node in enumerate(summary_response.source_nodes):
        print(f"소스 {i+1}: {node.text[:100]}...")
        if hasattr(node, 'score'):
            print(f"유사도 점수: {node.score:.3f}")
'''

print("\n" + "=" * 55)
print("\n💡 사용 가능한 명령어:")
print("- 질문을 입력하세요 (예: '주요 개념이 뭐야?')")
print("- 'debug': 마지막 답변의 소스 확인")
print("- 'history': 대화 기록 보기")
print("- 'summary': 강의 요약 다시 보기")
print("- 'check_doc': 문서 내용 일부 확인")
print("- 'exit': 종료")
print("-" * 55)

# 반복 제거 함수
def clean_response(text):
    """반복된 문장 제거"""
    sentences = text.split('.')
    unique_sentences = []
    seen = set()

    for sentence in sentences:
        sentence = sentence.strip()
        if sentence and sentence not in seen and len(sentence) > 10:
            unique_sentences.append(sentence)
            seen.add(sentence)

    return '. '.join(unique_sentences)

# 질의응답 루프
history = []
last_response = None

def show_source_info(response):
    """응답의 소스 정보를 표시하는 함수"""
    if hasattr(response, 'source_nodes') and response.source_nodes:
        print("\n🔍 **참조된 문서 부분:**")
        for i, node in enumerate(response.source_nodes):
            print(f"\n소스 {i+1}:")
            print(f"내용: {node.text[:200]}...")
            if hasattr(node, 'score'):
                print(f"유사도: {node.score:.3f}")
    else:
        print("\n⚠️ 소스 정보를 찾을 수 없습니다.")

while True:
    user_input = input("\n❓ 질문을 입력하세요: ").strip()

    if user_input.lower() == "exit":
        print("👋 질의응답을 종료합니다!")
        break

    elif user_input.lower() == "debug":
        if last_response:
            show_source_info(last_response)
        else:
            print("⚠️ 아직 질문이 없습니다.")

    elif user_input.lower() == "summary":
        print("\n📋 강의 요약:")
        print(clean_response(summary_response.response))
        show_source_info(summary_response)

    elif user_input.lower() == "history":
        print("\n📜 대화 기록:")
        for i, (q, a) in enumerate(history):
            print(f"{i+1}. Q: {q}")
            print(f"   A: {a[:200]}...")
            print()

    elif user_input.lower() == "check_doc":
        print("\n📄 문서 내용 확인:")
        if nodes:
            print(f"첫 번째 노드 내용: {nodes[0].text[:300]}...")
        else:
            print("노드가 없습니다.")

    elif user_input:
        # 질문 앞에 문서 기반 답변 강제하는 프리픽스 추가
        enhanced_query = f"문서 내용을 바탕으로만 답변해주세요: {user_input}"
        response = query_engine.query(enhanced_query)
        last_response = response

        # 반복 제거 처리
        cleaned_answer = clean_response(response.response)

        history.append((user_input, cleaned_answer))
        print("📝 **답변:**")
        print(cleaned_answer)

        # 소스 정보 표시
        # show_source_info(response)

    else:
        print("❗ 질문을 입력해주세요.")

# 추가 디버깅 함수들
def test_retrieval(query, top_k=3):
    """검색 기능 테스트"""
    retriever = index.as_retriever(similarity_top_k=top_k)
    nodes = retriever.retrieve(query)
    print(f"\n🔍 '{query}' 검색 결과:")
    for i, node in enumerate(nodes):
        print(f"\n결과 {i+1}:")
        print(f"내용: {node.text[:200]}...")
        if hasattr(node, 'score'):
            print(f"점수: {node.score:.3f}")

# 사용 예시:
# test_retrieval("주요 내용")
```
